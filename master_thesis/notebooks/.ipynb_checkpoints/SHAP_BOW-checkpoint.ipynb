## vergleichbar zu bei BERT: selber values speichern und betrachten!

```python
df_shap = pd.DataFrame(index=feature_names, columns = ["count", "mean", "mean_abs", "min", "max", "std"])
```

```python
df_shap.head()
```

```python
df_shap.shape
```

```python
# getting mean shap_value and count of the frequent tokens
for nr_feature, feature in enumerate(feature_names):
    print(nr_feature, feature, end="\r")
    values = shap_values[:, nr_feature] # all the shap values for the token
    count = np.sum(X_dev[:, nr_feature]) # das zählt Vorkommen (also auch mehrmals im gleichen Text)
    mean = np.mean(values)
    mean_abs = np.mean(abs(np.array(values)))
    min_value = np.min(values)
    max_value = np.max(values)
    std = np.std(values)
    df_shap.loc[feature]["mean"] = mean.round(3)
    df_shap.loc[feature]["mean_abs"] = mean_abs.round(3)
    df_shap.loc[feature]["count"] = count
    df_shap.loc[feature]["min"] = min_value.round(3)
    df_shap.loc[feature]["max"] = max_value.round(3)
    df_shap.loc[feature]["std"] = std.round(3)

    #if nr_feature >=10:
    #    break
```

```python
df_shap["abs"] = abs(df_shap["mean"])
```

```python
def model_predict(data):
    if isinstance(data, list):        
        X = vectorizer.transform(data)
        X = X.toarray()
        outputs = model.predict(X)
    if isinstance(data, np.ndarray):
        outputs = model.predict(data)
    return np.array(outputs)
```

```python
def scatter(true, pred):
    print(len(true), true[:5])
    print(len(pred), pred[:5])
    plt.scatter(true, pred, alpha = 0.1)
    plt.xlabel('True')
    plt.ylabel('Predictions')
    plt.axis('equal')
    plt.axis('square')
    plt.xlim(0,1000)
    plt.ylim(0,1000)
    plt.show()
    
    print("Pearson:", st.pearsonr(true, pred))
    print("Spearman:", st.spearmanr(true, pred)[0])

    print("MAE:", mean_absolute_error(true, pred))
    print("MSE:", mean_squared_error(true, pred))
    print("RAE:", utils.relative_absolute_error(pred, true))
```
